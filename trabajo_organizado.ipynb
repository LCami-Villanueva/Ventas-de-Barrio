# CLASE ZONA

```python
class Zona(Interfaz):

  # Tomamos los datos de la matriz de datos
  def tomar_datos(self,matriz,fila):
          self.nivel_del_mar = matriz[fila][0]
          self.area = matriz[fila][1]
          self.poblacion = matriz[fila][2]
          self.distancia_a_rio = matriz[fila][3]

  #determinamos si la poblacion es rural o urbana.
    # Según se considera población urbana a un núcleo de viviendas con un mínimo de 2.000 habitantes
  def determinar_urbana_o_rural(self):
    if self.poblacion > 1999:
      tipo_de_zona = "urbano"
    else:
      tipo_de_zona = "rural"
    self.tipo_de_zona = tipo_de_zona


  # determinamos si la zona es costera o montañosa.
    # Teniendo en consideración que el tsunami más alto registrado albergó olas de hasta de 524 metros de altura, asumiremos que el terreno montañoso inicia a partir de los 500 metros.
  def determinar_costa_o_montaña(self):
      if self.nivel_del_mar < 500:
          topografia = "costera"
      else:
          topografia = "montañosa"
      self.topografia = topografia


  # determinamos si existe riego y en caso de existir cuales son.
  def determinar_riesgo(self):
    tipos_de_riesgo = []

    if (self.topografia == "montañosa" and (self.distancia_a_rio <50)):
      self.existe_riesgo == "Sí"
      tipos_de_riesgo.append("Fluvial")

    if (self.topografia == "costera" and self.nivel_del_mar < 10):
      self.existe_riesgo = "Sí"
      tipos_de_riesgo.append("Costera")

    if ((self.poblacion/self.area) > 100):
      self.existe_riesgo = "Sí"
      tipos_de_riesgo.append("Urbana")  # mostramos los datos en la terminal
    
    if (len(tipos_de_riesgo)!= 0):
      self.tipos_de_riesgo = tipos_de_riesgo

    else:
      self.tipos_de_riesgo = "No"
    
  
  # Mostramos los datos.
  def mostrar_datos(self):
    print(f"Nivel del mar: {self.nivel_del_mar} m")
    print(f"Área: {self.area} km^2")
    print(f"Población: {self.poblacion} habitantes")
    print(f"Distancia al río: {self.distancia_a_rio} m")
    print(f"Topografía: {self.topografia}")
    print(f"Tipo de zona: {self.tipo_de_zona}")
    print(f"Existe riesgo: {self.existe_riesgo}")
    print(f"Tipos de Riego que existen: {self.tipos_de_riesgo}")



  # regresamos los datos cálculados a la matriz
  def regresar_datos(self,matriz,fila):
        matriz[fila][4] = self.tipo_de_zona
        matriz[fila][5] = self.topografia
        matriz[fila][6] = self.existe_riesgo
        
        if (self.tipos_de_riesgo == "No"):
          matriz[fila][7] = "No existen"

        else: 
          matriz[fila][7] = self.tipos_de_riesgo

  
```
